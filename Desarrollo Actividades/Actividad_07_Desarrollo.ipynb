#Desarrollo
Nuestro primera forma de desarrollar esta actividad es utilizando MLP. Luego de eso utilizamos train_test_split para generar el 70% de entrenamiento y 30% de prueba. A su vez utilizaremos multipples modelos con Relu, relu6 y swish para ver cual de ellas nos entrega el mejor accuracy.
# 1° Dos Layers, 128 neuronas cada uno, con 3 epocas - Relu

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import numpy as np
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x = np.concatenate((x_train, x_test))
y = np.concatenate((y_train, y_test))
#data de 0,7 entrenamiento y 0,3 de test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

#ESTO NO ES NECESARIO, PERO SIRVE PARA TENER UN MEJOR RESULTADO
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation = tf.nn.softmax))

model.compile(optimizer= 'adam',loss='sparse_categorical_crossentropy', metrics = ['accuracy'])
model.fit(x_train,y_train,epochs=3)

print("Resultados del modelo")
val_loss, val_acc = model.evaluate(x_test,y_test)
print('perdida:',round(val_loss,4),'accuracy', round(val_acc,4))
from sklearn.metrics import confusion_matrix
y_pred=model.predict(x_test)
arreglo=[]
for punto in y_pred:
  arreglo.append(np.argmax(punto))

y_pred=np.array(arreglo)
confusion_matrix(y_test, y_pred)
```

Con esto claro pasaremos a utilizar una red neuronal convolucional para ver los resultados que arroja y ver si podemos acercarnos al 0,99.
#1° Convolucional. Dos layers de Relu y una final de softmax
Antes de entregarle los datos debemos cambiar los datos para que se puedan utilizar en CNN.


```python
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D

train_x = x_train.reshape(-1, 28, 28, 1)
test_x = x_test.reshape(-1, 28, 28, 1)
train_y=y_train
modeloCNN = tf.keras.models.Sequential()
modeloCNN.add(Conv2D(150, (3, 3), input_shape=train_x.shape[1:]))
modeloCNN.add(Activation('relu'))
modeloCNN.add(MaxPooling2D(pool_size=(2, 2)))
modeloCNN.add(Conv2D(75, (3, 3)))
modeloCNN.add(Activation('relu'))
modeloCNN.add(MaxPooling2D(pool_size=(2, 2)))
modeloCNN.add(Flatten())
modeloCNN.add(Dense(64))
modeloCNN.add(Activation('softmax'))

modeloCNN.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
modeloCNN.fit(train_x, train_y, epochs = 8)


modeloCNN.summary()
```

```python
print("Resultados del modelo","/n")
val_loss, val_acc = modeloCNN.evaluate(test_x,y_test)
print('perdida:',round(val_loss,4),'accuracy',round(val_acc,4))
```

Con esto podemos notar como la Red convolucional es mucho más efectiva con los parametros que utilizamos. A su vez podemos ver que a medida que aumentan las epocas de ejecución el accuracy se acercará más y más a uno, por lo que esta red neuronal es mucho más precisa que la MLP.
